```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris_data = load_iris()
```

```python
# Create a DataFrame
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target
iris['species'] = iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
print(iris.head())
```

```python
print(iris.describe())
```

```python
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.show()
```

```python
# Select only the numerical columns before calculating correlation
numerical_iris = iris.select_dtypes(include=['float', 'int'])
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_iris.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal length (cm)', data=iris)
sns.violinplot(x='species', y='sepal width (cm)', data=iris)
sns.violinplot(x='species', y='petal length (cm)', data=iris)
sns.violinplot(x='species', y='petal width (cm)', data=iris)
plt.show()
```

```python
iris.to_csv('iris_dataset.csv', index=False)
```
